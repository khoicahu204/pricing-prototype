```python
import seaborn as sns
import matplotlib.pyplot as plt

df_corr = df[['price', 'manufacture_date', 'mileage_v2', 'seats']].copy()
df_corr['brand'] = df['brand'].astype('category').cat.codes
df_corr['type'] = df['type'].astype('category').cat.codes
df_corr['model'] = df['model'].astype('category').cat.codes
corr_matrix = df_corr.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Heatmap tương quan giữa các biến")
plt.tight_layout()
plt.show()
```
